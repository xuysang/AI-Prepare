```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# 准备数据
x = np.arange(1,6)
y = np.array([3,7,1,5,9])
df = pd.DataFrame({
    "day":["Mon", "Tue", "Wed", "Thu", "Fri"],
    "sales":[120,90,100,140,160]
})

# ==============================
# 1. Matplotlib 基础
# ==============================
plt.Figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x,y,marker='o',color='blue',label="y values")
plt.title("折线图plot")
plt.xlabel("x轴")
plt.ylabel("y轴")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 柱状图
plt.subplot(2,2,2)
plt.bar(df["day"],df["sales"],color="orange")
plt.title("柱状图 Bar")
plt.grid(axis='y')
plt.show()
```

```python
# 直方图
plt.subplot(2,2,3)
data = np.random.randn(1000)
plt.hist(data,bins=20,color="green",edgecolor="black")
plt.title("直方图 Hist")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# ==============================
# 2. Seaborn 基础
# ==============================
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.lineplot(data=df,x="day",y="sales",marker='o',color='red')
plt.title("Seaborn 折线图")
plt.show()
```

```python
plt.subplot(1,2,2)
sns.barplot(data=df,x="day",y="sales",hue="day",palette="viridis",legend=False)
plt.title("Seaborn 柱状图")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.barplot(data=df,x="day",y="sales",color="skyblue")
plt.title("保存为PNG")
plt.savefig("sales_bar.png",dpi=300)
plt.close()
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.title('Sine Function', fontsize=14)
plt.xlabel('x (radians)', fontsize=12)
plt.ylabel('sin(x)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7) # 显示网格

# 显示图形
plt.tight_layout() # 自动调整布局
plt.show()
```
